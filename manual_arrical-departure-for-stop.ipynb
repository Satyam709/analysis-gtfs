```python
stop26080Trips =  stopsTime[stopsTime["stop_id"] == 26080].sort_values(by="arrival_time", inplace=False)
```

```python
# Step 1: Filter stop_times for stop_id 26080
stop26080Trips = stopsTime[stopsTime["stop_id"] == 26080].sort_values(by="arrival_time", inplace=False)

# Step 2: Merge with trips to get service_id
stop26080Trips = stop26080Trips.merge(trips[['trip_id', 'service_id']], on='trip_id', how='left')

# Step 3: Merge with calendar to get day-of-week activity
stop26080Trips = stop26080Trips.merge(calendar, on='service_id', how='left')

# Step 4 (Optional): Select and reorder relevant columns
days_columns = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
cols_to_show = ['trip_id', 'arrival_time', 'departure_time', 'stop_id'] + days_columns

stop26080Trips = stop26080Trips[cols_to_show]

stop26080Trips.to_csv("./data/arrival_departures_for_stop26080Trips.csv", index=False)
```
